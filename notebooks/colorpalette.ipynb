```python
from matplotlib.colors import LinearSegmentedColormap
import numpy as np
from mapio.shake import ShakeGrid
from mapio.gdal import GDALGrid
import matplotlib.pyplot as plt

from impactutils.colors.cpalette import ColorPalette
```

```python
#make a test map
shakefile = '/Users/mhearne/data/shakemaps/northridge.xml'

shakegrid = ShakeGrid.load(shakefile,adjust='res')

mmidata = shakegrid.getLayer('mmi').getData()
cp = ColorPalette.fromPreset('mmi')
cp._cdict
plt.figure(figsize=(10,8))
plt.imshow(mmidata,vmin=cp.vmin,vmax=cp.vmax,cmap=cp.cmap)
plt.colorbar();
```

```python
popfile = '/Users/mhearne/pager/data/lspop2012.flt'
popgrid = GDALGrid.load(popfile,samplegeodict=shakegrid.getGeoDict())
popdata = popgrid.getData()
cp2 = ColorPalette.fromPreset('pop')
plt.figure(figsize=(10,8));
plt.imshow(popdata,vmin=cp2.vmin,vmax=cp2.vmax,cmap=cp2.cmap)
plt.colorbar();
```
